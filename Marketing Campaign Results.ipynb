```python
#Import the required Libraries
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

#loading the data
data_describe = pd.read_csv('marketing_data_dictionary.csv')
data = pd.read_csv('marketing_data.csv')

#copying the dataset so that we can use original dataset when needed
data_original = data
```

```python
# calculate age and replace with year of birth
current_year = datetime.now().year
data['Age'] = current_year - data['Year_Birth'] 


# customer duaration(customer stay) with year_num
years = (data['Dt_Customer'].str.split('-')).apply(lambda p:p[0])
years = years.astype(int)
duration = current_year - years 
data['Duration'] = duration

# now drop the 'Year_Birth' and 'Dt_Customer' columns
data = data.drop(['Year_Birth','Dt_Customer'],axis = 1)      #run only once
```

```python
data.columns
```

```python
#like we can see there is a space in the income col name i.e " income "
#lets rename the column
data=data.rename(columns={' Income ': 'Income'})
```

```python
# Its time to remove the outliers
rows = data.loc[data['Income'] > 200000]
data = data.drop(rows.index,axis = 0)
```

```python
#Adding the amount columns and purchase columns
data['amtTotal'] = data['MntWines'] + data['MntFruits'] + data['MntMeatProducts'] + data['MntFishProducts'] + data['MntSweetProducts'] + data['MntGoldProds']
data['numPurchases'] = data['NumDealsPurchases'] + data['NumWebPurchases'] + data['NumCatalogPurchases'] + data['NumStorePurchases']
```

```python
# lets put a range on income for better visualization
baskets = list(range(0,150000,30000))
data['Income_bin'] = pd.cut(data['Income'],baskets)
```

```python
data['Income_bin'].head(10)
```

```python
""" 1. What factors are significantly related to the number of web purchases? """


#Lets first check Web Purchases with respect to education
labels = ['Basic','2n Cycle','Graduation','PhD','Master']
educ_list = [data.loc[data['Education'] == label, 'NumWebPurchases'] for label in labels]

colors = ['blue', 'yellow', 'cyan', 'hotpink', 'lightgreen']
plt.hist(educ_list, label=labels, histtype='bar', stacked=True, bins=25, color=colors)
plt.legend()
plt.xlabel('NumWebPurchases')
plt.ylabel('Frequency')
plt.title('Distribution of NumWebPurchases by Education Level')
plt.show()
```

```python
"""
Customers with higher education tend to place more web purchases,
thus education is one of the factors that influences Number of purchases
"""
```

```python
#web purchases with respect to Country
grouped = data.groupby('Country')['NumWebPurchases'].sum()

plt.pie(grouped,labels = grouped.index,startangle=90, colors=plt.cm.Paired.colors)

plt.title('Total NumWebPurchases by Country')
plt.show()
```

```python
"""
Significant number, almost half of Web purchases are from Spain,
followed by Saudi, geography is a significant factor
"""
```

```python
# NumWebPurchases with respect to Income
# Group by 'Income_bin' and sum the 'NumWebPurchases'
grouped = data.groupby('Income_bin')['NumWebPurchases'].sum()

# Define the order of income bins
x = ['0-30k', '30k-60k', '60k-90k', '90k-120k']

# Extracting y values from the grouped data
y = grouped.values

plt.bar(x, y, color='purple')

plt.xlabel('Income')
plt.ylabel('Web Purchases')
plt.title('Effect of Income on Web Purchases')

# Show the bar chart
plt.show()
```

```python
"""
Middle class(30k-60k) and upper middle class(60k-90k) people tend to order more on
Web. Lower class(0-30k) and upper class(90k-120k) don't order much through 
internet.
"""
```

```python
# trying to find effect of other factors

fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize = (10,10))


# Marital_Status
grouped_data = data.groupby('Marital_Status')['NumWebPurchases'].sum()
ax1.pie(grouped_data, labels = grouped_data.index )
ax1.set_title('Effect of Marital Status on Web Purchase')

# Kidhome
grouped_data = data.groupby('Kidhome')['NumWebPurchases'].sum()
ax2.pie(grouped_data, labels = grouped_data.index )
ax2.set_title('Effect of kids in home on Web Purchase')

# Teenhome
grouped_data = data.groupby('Teenhome')['NumWebPurchases'].sum()
ax3.pie(grouped_data, labels = grouped_data.index )
ax3.set_title('Effect of teens in home on Web Purchase')

fig.delaxes(ax4)
plt.show()
```

```python
""" 2. Which marketing campaign was the most successful? """

# Customers who responded to campaigns 1, 2, 3, 4, 5, and last campaign
cmp1 = data.loc[data.AcceptedCmp1 == 1]
cmp2 = data.loc[data.AcceptedCmp2 == 1]
cmp3 = data.loc[data.AcceptedCmp3 == 1]
cmp4 = data.loc[data.AcceptedCmp4 == 1]
cmp5 = data.loc[data.AcceptedCmp5 == 1]
last = data.loc[data.Response == 1]
not_responded = data.loc[data.Response == 0]

x = ['cmp1', 'cmp2', 'cmp3', 'cmp4', 'cmp5', 'last']
y = [len(cmp1), len(cmp2), len(cmp3), len(cmp4), len(cmp5), len(last)]

plt.bar(x, y, color='green')
plt.xlabel('Campaign')
plt.ylabel('Customers')
plt.title('Campaigns Well Received by Customers')
plt.show()
```

```python
# From the above plot it's clear that majority of people responded to last Campaign
# Now let's check which campaign resulted highest revenue

y = [cmp1['amtTotal'].sum(), cmp2['amtTotal'].sum(), cmp3['amtTotal'].sum(),
     cmp4['amtTotal'].sum(), cmp5['amtTotal'].sum(), last['amtTotal'].sum()]

x = ['cmp1', 'cmp2', 'cmp3', 'cmp4', 'cmp5', 'last']

plt.bar(x, y, color='red')
plt.xlabel('Campaign')
plt.ylabel('Amount')
plt.title('Amount generated by the Campaigns')
plt.show()
```

```python
"""
Overall, last campaign run by the company is the most successful among all
First and fifth campaigns have performed well standing next to the last campaign.
"""
```
